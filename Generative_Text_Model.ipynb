```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Step 1: Load Pre-trained GPT-2 Model and Tokenizer
model_name = "gpt2-medium"  # You can also use 'gpt2' or 'gpt2-large'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("Model loaded successfully!")

# Step 2: User Input for Prompt
prompt_text = input("Enter a prompt for the text generator: ")

# Encode the input prompt
inputs = tokenizer.encode(prompt_text, return_tensors="pt").to(device)

# Step 3: Generate Text
print("\nGenerating text... Please wait...\n")
output = model.generate(
    inputs,
    max_length=200,          # Maximum length of the generated text
    num_return_sequences=1,  # Number of sequences to generate
    temperature=0.7,         # Sampling temperature; lower is more focused, higher is more random
    top_k=50,                # Top-K sampling
    top_p=0.95,              # Nucleus sampling
    do_sample=True,          # Random sampling enabled
    early_stopping=True
)

# Step 4: Decode and Display the Output
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Generated Text:\n")
print(generated_text)
```
